Getting data from github

```python
import requests
import pandas as pd
```

```python
#git raw link

url = 'https://raw.githubusercontent.com/WahuElizabethKE/synthetic_employees/refs/heads/main/MOCK_DATA%20(2).json'
```

```python
#fetching data from Github

response = requests.get(url)
```

```python
# Check if request was successful
if response.status_code == 200:
    print("✅ Successfully fetched data from GitHub!")
else:
    print("❌ Failed to fetch data. Check your URL.")
    exit()
```

```python
#Convert JSON to Python list/dictionary
data = response.json()
```

```python
#Convert to Pandas DataFrame
employee_df = pd.DataFrame(data)
```

```python
#Display info
print(f"Dataset Shape: {employee_df.shape}")
print("\nFirst Few Rows:")
print(employee_df.head())
```

```python
#replacinng deparment error with unknown
employee_df['department'] = employee_df['department'].replace("error: undefined method `first' for nil:NilClass", 'finance')
```

```python
#printing to see changes
print(employee_df['department'])
```

```python
employee_df['department'] = employee_df['department'].replace("Unknown", 'Finance')
```

```python
#changing uknown to finance

print(employee_df['department'])
```

```python
print(employee_df )
```

```python
employee_df[-5:]
```

```python
#Save the cleaned version

employee_df.to_csv("employees_clean.csv", index=False)
print("File saved as employees_clean.csv")
```

```python
#checking data type
# Python Lists vs. Dictionaries:
# Lists:
# - Ordered collection of items.
# - Items are accessed by their integer index (e.g., my_list[0]).
# - Can contain duplicate items.
# - Enclosed in square brackets [].
# Example: [1, 2, 3, 'apple', 'banana']

# Dictionaries:
# - Unordered collection of key-value pairs.
# - Items are accessed by their unique keys (e.g., my_dict['name']).
# - Keys must be unique and immutable (e.g., strings, numbers, tuples).
# - Values can be of any type and can be duplicated.
# - Enclosed in curly braces {} with key:value pairs.
# Example: {'name': 'Alice', 'age': 30, 'city': 'New York'}

# In your current notebook, 'phone_data' is a list of dictionaries.
# Each dictionary within the list represents a phone with keys like 'brand', 'year', 'price', etc.

type(phone_data)
```
